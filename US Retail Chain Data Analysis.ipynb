### **Let's start Anayse the Data**

#### First, we want to know the Total Sale

```python
retail['TOT_SALES'].sum().round(2)
```

#### Now, we want to know the Total Sale & Avg. Total Sale by Type of Customers

```python
retail.pivot_table(index=['PREMIUM_CUSTOMER'],values=['TOT_SALES'],aggfunc=['sum','mean'])
```

#### Here we found Mainstream has the Maximum Sale.

#### And now, we want to know the Total sale & Avg. Total Sale by Lifestage or Age group of customers

```python
retail.pivot_table(index=['LIFESTAGE'],values=['TOT_SALES'],aggfunc=['sum','mean'])
```

#### Here we found Older Singles/Couples has the Maximum Sale and New Families has the Minimum Sale.

#### Let's check Total Sale by Agegroup and Type of Customers together

```python
retailgrp=retail.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TOT_SALES'].sum().unstack()
```

```python
retailgrp
```

#### Now, create a Plot to better understand the result

```python
plt.figure(figsize=(20,5))
retailgrp.plot(kind='bar',stacked=True,color=['lightcoral','cyan','limegreen'])
plt.title('Total Sale by Customer Lifestage and Type of Customer')
```

#### We found by Plot, New Families has the Minimum Sale in each Customer Category.
#### Let's check Total Sale & Avg. Total Sale by Brands name

```python
retail.pivot_table(index=['BRAND'],values=['TOT_SALES'],aggfunc=['sum','mean'])
```

#### And here we found Kettle has Maximum Sale and Burger has Minimum Sale.
#### Create a plot for better understanding about Brands Sale

```python
# Create Bar plot using Seaborn
sns.barplot(data=retail,x='BRAND',y='TOT_SALES',estimator=sum)
plt.xticks(rotation=90)
```

#### Create Bar plot using Pandas plot function and Groupby (same records and fields)

```python
retailbrand=retail.groupby(['BRAND'])['TOT_SALES'].sum()
```

```python
retailbrand.plot(kind="bar")
plt.title("Total Sale by Brands")
```

#### Here we can see the difference between Kettle and Burger Brands.
#### Now, Let's check Total Sale by Brands & Products Name

```python
# Using groupby Unstacked Table
bpngrp=retail.groupby(['PROD_NAME','BRAND'])['TOT_SALES'].sum().unstack()
bpngrp
```

#### Check the Total_Sale by Brands and Product Nmae usig Pivot Table (same fields)

```python
bpnpvt=retail.pivot_table(index=['BRAND','PROD_NAME'],values=['TOT_SALES'],aggfunc=['sum'])
bpnpvt
```

#### and by this result we found Product- Dorito Corn Chip Supreme 380g has Maximum Sale and Woolworths Mild Salsa 300g has Minimum Sale.
### Now, for analysis with Dates we have to extract Day, Month and Year from DATE column.

```python
retail.head()
```

#### Check the datatype of all the Fields

```python
retail.dtypes
```

#### Here, we can see DATE column has Object datatype. So, we have to convert this column into Date format

#### Convert DATE column into Datetime format

```python
retail=retail.astype({'DATE':'datetime64[ns]'})
retail.dtypes
```

#### Now, Let's extract the Month and Year from DATE column

```python
retail['year']=retail['DATE'].dt.year
retail['month']=retail['DATE'].dt.month
```

```python
retail.head()
```

#### Let's check Total Sale by Year

```python
retail.pivot_table(index=['year'],values=['TOT_SALES'],aggfunc=['sum'])
```

#### And here we found 2018 (July to December) has more Sale than 2019 (January to June).
#### Let's deep dive into Year with Months

```python
retail.pivot_table(index=['year','month'],values=['TOT_SALES'],aggfunc=['sum'])
```

#### From this result we found December has the Maximum Sale and February has Minimum Sale.
#### Let's create Plot to understand more about Sale (Date wise)

```python
plt.figure(figsize=(25,10))
sns.lineplot(data=retail,x='DATE',y='TOT_SALES',estimator='sum')
```
